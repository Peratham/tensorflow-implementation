```python
import tensorflow as tf
import pandas as pd
import os
import numpy as np
import json
import pickle
from BeautifulSoup import BeautifulSoup
```

```python
annotation_path = '../../../Public/Image_Dataset/train2014_annotations/captions_train2014.json'
image_path = "../../../Public/Image_Dataset/train2014_resized/"
annotation_result_path = './mscoco_pickles/annotations.pickle'
```

```python
f = json.loads(open(annotation_path).read())
image_list = []
for i in range(len(f['images'])):
    image_list.append(f['images'][i]['id'])
```

```python
data = []
for i in range(len(image_list)):
    for item in f['annotations']:
        if item['image_id'] == image_list[i]:   
            item['file_name'] = f['images'][i]['file_name']
            data += [item]
    if i  % 500 == 0:
        print str(i) + " finish"      
captions = pd.DataFrame.from_dict(data)
captions.head()
```

```python
captions.to_pickle(annotation_result_path)
```
