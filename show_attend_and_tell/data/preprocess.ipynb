```python
import cPickle as pickle
import numpy as np
```

```python
with open('annotations.pickle') as f:
    annotations = pickle.load(f)
```

```python
with open('features.pickle') as f:
    features = pickle.load(f)
```

```python
idx_drop = []
for i, sentence in enumerate(annotations['caption']):
    sentence = sentence.replace('.','').replace(',','').replace("'","").replace('"','')

    if len(sentence.split(" ")) > 20:
        idx_drop.append(i)

annotations = annotations.drop(annotations.index[idx_drop])
```

```python
captions = annotations['caption']
```

```python
def build_word_to_idx(sentences, threshold=200): 

    word_counts = {}
    n_sents = 0
    max_len = 0
    for sentence in sentences:
        sentence = sentence.replace('.','').replace(',','').replace("'","").replace('"','')

        if len(sentence.split(" ")) > max_len:
            max_len = len(sentence.split(" "))

        n_sents += 1
        for word in sentence.lower().split(' '):
            word_counts[word] = word_counts.get(word, 0) + 1
    vocab = [word for word in word_counts if word_counts[word] >= threshold]
    print 'Filtered words from %d to %d' % (len(word_counts), len(vocab))

    word_to_idx = {u'<NULL>': 0, u'<START>': 1, u'<END>': 2, u'<UNK>': 3}
    idx_to_word = {0: u'<NULL>' , 1: u'<START>', 2: u'<END>', 3:u'<UNK>'}
    
    
    idx = 4
    for word in vocab:
        word_to_idx[word] = idx
        idx_to_word[idx] = word
        idx += 1

    return word_to_idx, idx_to_word, max_len

```

```python
word_to_idx, idx_to_word, max_len = build_word_to_idx(captions)
print "max length of caption: ", max_len
```
